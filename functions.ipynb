**Question no. 6. Explain the concept of generators in Python and  how they are defined.**

**Answer.** A generator is a special kind of iterator.

It generates values on the call instead of storing everything in memory.

It uses the yield keyword (instead of return).

```python
#defining a generator >> it is similar to defining a function but instedad of return we will use yield.
def count_up_to_three():
    yield 1
    yield 2
    yield 3
```

```python
gen = count_up_to_three()
print(next(gen))  # 1
```

```python
print(next(gen))  # 2

```

```python
print(next(gen))  # 3
```

**Question no. 10. What is the difference between 'map()',  'reduce()', and 'filter()' functions in Python.**

**Answer.** Map() executes an specified function for each of the iteml in an iterable.

Reduce() is used for folding or reduction of the function.
it will always take 2 arguments function and if the iterable is empty it will throw an error. We have to import it from functools.

```python
#map
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x*x, numbers))
print(squared)  # Output: [1, 4, 9, 16]
```

**Question no. 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**


```python
def sum_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return sum_even
```

```python
my_list = [1, 2, 45, 68, 3, 49, 54, 63, 76]
result = sum_even_numbers(my_list)
print(result)
```

**Question no. 4. Write  a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
```

**Question no. 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return next_value
```

```python
fib = FibonacciIterator(25)
for num in fib:
    print(num)
```

**Question no. 6. Write a generator function in python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
```

```python
for value in powers_of_two(5):
    print(value)
```

**Question no. 7. Implement  a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
#for example file named "statistics.txt"
for line in read_file('statistics.txt'):
    print(line)
```

**Question no. 9. Write a Python program that uses 'map()' to convert a list of temperatures from Celsius to Fahrenheit.**

```python
celsius_temps = [0, 10, 20, 30, 40, 45, 67, 37]

# Formula to convert Celsius to Fahrenheit: (C × 9/5) + 32
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print(fahrenheit_temps)
```

**Question no. 10. Create a Python program that uses 'filter()' to remove all the  vowels from a given string.**

```python
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    # Keep only characters that are NOT vowels
    result = ''.join(filter(lambda ch: ch not in vowels, text))
    return result
```

```python
string = "Create a Python program that uses 'filter()' to remove all the vowels from a given string."
filtered_string = remove_vowels(string)
print(filtered_string)
```
